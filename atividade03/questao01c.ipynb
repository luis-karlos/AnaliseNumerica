## **Questão 1 - c)**
$\begin{cases}
  x' = -\frac{1}{2}x + y \\
  y' = -x - \frac{1}{2}y \\
  x(0) = 1, y(0) = 1
\end{cases}$


```python
def sol_exata(t):
    x = np.exp(-t/2)*(np.cos(t) + np.sin(t))
    y = np.exp(-t/2)*(np.cos(t) - np.sin(t))
    return np.array([x, y])

def sistema(t, U):
    x, y = U
    dx = (-1/2)*x + y
    dy = -x - (1/2)*y
    return np.array([dx, dy])

T = 15
dt = [.1, .01]
ponto_critico = [0, 0]
ponto_inicial = [1, 1]
tempo_inicial = 0
```

```python
plt.figure(figsize=(12, 7))
y_exato = sol_exata(np.linspace(0, T, 100))
plt.plot(y_exato[0, :], y_exato[1, :], 'k-', label='Solução exata')
for h in dt:
    _, y_euler = m.metodo_euler_explicito(sistema, [tempo_inicial, ponto_inicial], T, h)
    plt.plot(y_euler[:, 0], y_euler[:, 1], '--', label=f'Euler h={h}')
plt.plot(ponto_inicial[0], ponto_inicial[1], 'bx')
plt.plot(ponto_critico[0], ponto_critico[1], 'rx')
plt.ylabel('y(t)')
plt.xlabel('x(t)')
plt.legend()
nome_arquivo = f"{diretorio_imagem}/euler.pdf"
plt.savefig(nome_arquivo)
print(f"✅ Gráfico salvo com sucesso no arquivo: {nome_arquivo}")
plt.show()
```
