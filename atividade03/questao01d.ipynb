```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import ipynbname
import mylib as m
import scienceplots

plt.style.use(['grid'])
diretorio_imagem = f"img/{ipynbname.name()}"
os.makedirs(diretorio_imagem, exist_ok=True)
```

## **Questão 1 - d)**
$\begin{cases}
  x' = -3x +2z \\
  y' = x - y \\
  z' = -2x - y \\
  x(0) = 1, y(0) = 1, z(0) = 1
\end{cases}$


```python
def sol_exata(t):
    x = -(2/3)*np.exp(-2*t) + np.exp(-t)*((5/3)*np.cos(t*np.sqrt(2))-(np.sqrt(2)/3)*np.sin(t*np.sqrt(2)))
    y = (2/3)*np.exp(-2*t) + np.exp(-t)*((5*np.sqrt(2)/6)*np.sin(t*np.sqrt(2))+(1/3)*np.cos(t*np.sqrt(2)))
    z = -(1/3)*np.exp(-2*t) + np.exp(-t)*((4/3)*np.cos(t*np.sqrt(2)) - (7*np.sqrt(2)/6)*np.sin(t*np.sqrt(2)))
    return np.array([x, y, z])

def sistema(t, U):
    x, y, z = U
    dx = -3*x + 2*z
    dy = x - y
    dz = -2*x - y
    return np.array([dx, dy, dz])
T = 10
dt = [.1, .01]
ponto_critico = [0, 0, 0]
ponto_inicial = [1, 1, 1]
tempo_inicial = 0
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10), constrained_layout=True)
y_exato = sol_exata(np.linspace(0, T, 100))

ax[0, 0].plot(y_exato[0, :], y_exato[1, :], 'k-', label='Solução exata')
ax[0, 0].plot(ponto_inicial[0], ponto_inicial[1], 'bx')
ax[0, 0].plot(ponto_critico[0], ponto_critico[1], 'rx')

ax[0, 1].plot(y_exato[1, :], y_exato[2, :], 'k-', label='Solução exata')
ax[0, 1].plot(ponto_inicial[1], ponto_inicial[2], 'bx')
ax[0, 1].plot(ponto_critico[1], ponto_critico[2], 'rx')

ax[1, 0].plot(y_exato[0, :], y_exato[2, :], 'k-', label='Solução exata')
ax[1, 0].plot(ponto_inicial[0], ponto_inicial[2], 'bx')
ax[1, 0].plot(ponto_critico[0], ponto_critico[2], 'rx')

for h in dt:
    _, y_euler = m.metodo_euler_explicito(sistema, [tempo_inicial, ponto_inicial], T, h)
    
    ax[0, 0].plot(y_euler[:, 0], y_euler[:, 1], '--', label=f'Euler h={h}')
    ax[0, 0].set_xlabel('x(t)')
    ax[0, 0].set_ylabel('y(t)')
    ax[0, 0].set_aspect('equal', adjustable='box')
    ax[0, 0].set_box_aspect(1)
    ax[0, 0].legend()

    ax[0, 1].plot(y_euler[:, 1], y_euler[:, 2], '--', label=f'Euler h={h}')
    ax[0, 1].set_xlabel('y(t)')
    ax[0, 1].set_ylabel('z(t)')
    ax[0, 1].set_aspect('equal', adjustable='box')
    ax[0, 1].set_box_aspect(1)
    ax[0, 1].legend()

    ax[1, 0].plot(y_euler[:, 0], y_euler[:, 2], '--', label=f'Euler h={h}')
    ax[1, 0].set_xlabel('x(t)')
    ax[1, 0].set_ylabel('z(t)')
    ax[1, 0].set_aspect('equal', adjustable='box')
    ax[1, 0].set_box_aspect(1)
    ax[1, 0].legend()

for a in ax.flatten():
    a.set_xlim([-.75, 1.1])
    a.set_ylim([-.75, 1.1])
ax[1, 1].axis('off')

nome_arquivo = f"{diretorio_imagem}/euler.pdf"
plt.savefig(nome_arquivo)
print(f"✅ Gráfico salvo com sucesso no arquivo: {nome_arquivo}")
plt.show()
```
